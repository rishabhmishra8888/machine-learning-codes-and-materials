```python
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv("C://Users/mishr/Desktop/excel files/nyc_temperature.csv",nrows=10)
```

```python
df
```

```python
x=df.date
y=df.tmax
z=df.tmin
t=df.tavg
```

```python
plt.xlabel("date")
plt.ylabel('temprature')
plt.title("variations in temprature")
plt.plot(x,y,color="g",label="maximum temprature")
plt.plot(x,z,color='purple',label="minimum temprature")
plt.legend(loc ="lower right",shadow=True)
```

```python
plt.bar(x,t,width=0.8,label="average temprature")
plt.bar(x,y,width=0.2,label="max_temprature")
plt.ylabel("temprature")
plt.xlabel("DATE")
plt.title("WEATHER")
plt.legend()

```

```python
plt.barh(x,z)
```

```python
plt.hist(y,bins=[30,35,40,45,50,55,60,65],rwidth=1,color="purple",label="maximum temprature")
plt.hist(z,bins=[30,35,40,45,50,55,60,65],rwidth=0.5,color="red",label="minimum temprature")

plt.xlabel("temprature")
plt.ylabel("number of days")
plt.title("WEATHER")
plt.legend(loc="best",shadow=True)
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True,radius=2.3)
plt.axis("equal")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.3f%%',radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.3,0.2],counterclock=True, startangle=45)
plt.show()
```
