```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[22,27,34,21,16,28,26]
plt.xlabel("day")
plt.ylabel('temprature')
plt.title("variation in temprature")
plt.plot(x,y,color='green',linewidth=5,linestyle="dotted")
```

```python
X = np.linspace(0, 2*np.pi, 100)
Y = np.cos(X)
fig, ax = plt.subplots()
ax.plot(X, Y, color='green')
plt.show()
```

## MATPLOTLIB LECTURE 3
AXES LABEL LEGEND AND GRID

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("days")
plt.ylabel("temprature")
plt.title("variation in weather")
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="average")
plt.legend(loc="best",shadow=True,fontsize="medium")
plt.grid()
```
